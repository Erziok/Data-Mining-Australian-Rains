##1.2 Investigación (contextualización)

Para entender mucho mejor el caso, tuvimos que indagar e investigar cómo funcionan las precipitaciones, ya que no podemos predecir algo de lo que no se tiene conocimiento. Por esto hemos decidido realizar esta investigación que detalla los siguientes puntos:

* **Composición de las precipitaciones:**

>Las nubes se componen de hielo o micro gotas de agua dependiendo de la altura de la nube y la temperatura atmosférica.
>
>El vapor de agua que contiene el aire al enfriarse se condensa, formando las pequeñas gotas de agua que forman las nubes. Son tan livianas molecularmente que flotan en el aire (menos densas que el aire).

* **Medición de las precipitaciones:**

>Precipitación pluvial: Precipitación intensa pero de corta duración.
La forma más común de medirla es en MM que sería el espesor de la lámina de agua que se formaría a causa de la precipitación sobre una superficie plana e impermeable y que equivale a litros de agua por metro cuadrado del terreno (L/m2).

* **Origen de la precipitación:**

>Toda precipitación se produce por la condensación del vapor de agua contenido en las masas de aire que se originan cuando dichas masas de aire son forzadas a elevarse y enfriarse (cuando el aire caliente y húmedo cercano a la superficie terrestre asciende y su temperatura baja). Para que se produzca la condensación es preciso que el aire se encuentre saturado de humedad y existan núcleos de condensación.
>
>(El aire se satura si contiene muchas moléculas de vapor de agua). El estado de saturación se alcanza normalmente por el enfriamiento del aire. ya que el aire frío se satura con menor cantidad de vapor de agua que el aire caliente.
>
>Por ejemplo, un milímetro cúbico (m³) de aire a 25ºC, cuyo contenido de vapor de agua sea de 11g, no está saturado; pero los 11g lo saturan a 10ºC, y entonces la condensación (cambio de estado gaseoso a líquido) ya es posible 
>
>La pequeñez de las gotas y de los cristales les permite quedar en suspensión en el aire y ser desplazadas por los vientos. Se pueden contar 500 por cm³ y, sin embargo, 1 m³ de nube apenas contiene tres gramos de agua.

* **Variación temporal:**

>Variación de acuerdo a ciertos ciclos temporales determinados por los movimientos de rotación y traslación. Por ejemplo, en San Francisco, California (Estados Unidos), los meses de mayores precipitaciones se dan entre noviembre y marzo, mientras que en Miami, Florida los meses de mayor precipitación son de mayo a octubre.

* **Variación de la precipitación:**

>La distribución espacial de la precipitación sobre los continentes es muy variada, así existen extensas áreas como los desiertos, donde las precipitaciones son extremadamente escasas, del orden de 0 a 200 mm de precipitación por año. En el desierto del Sahara la media anual de lluvia es de apenas algunos mm, mientras que en las áreas próximas al golfo de Darién entre Colombia y Panamá, la precipitación anual es superior a 3000 mm, con un máximo de unos 10 metros (10.000 mm). El desierto de Atacama en el norte de Chile, es el área más seca de todos los continentes.
>
>De esto quisimos identificar los principales tipos de lluvias, del cual podemos destacar cuatro tipos: 

* **Precipitaciones por presión:**

>Cuando se junta una corriente de aire frío y cálido se crea en medio un área de bajas presiones, y la nubosidad se crea en áreas pequeñas, provocando que las micro gotas de agua aumenten su densidad más rápidamente.

* **Precipitaciones por convección:**

>Cuando la superficie de la tierra está muy caliente y hay viento frío en la troposfera, el aire cálido que pesa menos, asciende y al encontrarse con el aire frío produce cumulonimbos que dan lugar a lluvia acompañadas de tormenta (no suelen durar mucho). 

* **Precipitaciones por orografía:**

>Cuando el aire húmedo llega a las montañas debe elevarse más y eso provoca que se enfríe más, lo que provoca una condensación del vapor y formación de las nubes, si la masa aumenta mucho o se unen muchas masas pueden dar sitio a lluvias o nieve.

* **Precipitaciones por frente cálido:**

>Si se encuentran dos masas de aire, cálido y frío, la masa cálida se desplaza por encima de la fría, el límite de las dos masas da sitio a condensación y precipitaciones.

```python
muestra_o = datos_aus[["Date","Location","Temp9am","Temp3pm","Humidity3pm","Humidity9am","MinTemp","MaxTemp","Rainfall","RainToday","RainTomorrow"]]
muestra_o
```

Como anteriormente en el Data Understanding se analizó los datos nulos presentes en el set de datos, ahora es cuando debemos tratar dichos datos. Para este proceso contamos con 3 opciones de tratamiento:

- Eliminar los datos nulos o duplicados
- Trabajar con estos valores sin tratarlos
- Tratar de estimar estos valores ausentes utilizando el resto del set de datos

Nosotros como grupo hemos decicido tomar la desición de tratar de estimar los valores nulos dentro de la muestra. Esto utilizando los valores de la fecha para estimar el valor perdido en base al día siguiente y al día anterior. En el caso de RainFall y RainToday se haría esta misma estimación pero también fijandose en el valor de RainTomorrow del día anterior.

```python
muestra_o_sucia = muestra_o.sort_values('Date',ascending=True)
```

Al realizar el conteo de los datos nulos dentro de la muestra, podemos evidenciar la carencia de valores perdidos en las columnas Date, Location y RainTomorrow, las cuales serán de ayuda para la estimación de los valores nulos del resto de columnas.

```python
#Quitar datos nulos de raintoday
for i in muestra_o_sucia.index:
  if((muestra_o_sucia['RainToday'][i] != "Yes") & (muestra_o_sucia['RainToday'][i] != "No")):
    if(i == 0):
      muestra_o_sucia['RainToday'][i] = muestra_o_sucia['RainTomorrow'][i]
    else:
      muestra_o_sucia['RainToday'][i] = muestra_o_sucia['RainTomorrow'][i-1]

print("\033[4;32m"+'Los datos nulos de RainToday han sido imputados. DataFrame Actualizado!')
```

```python
#Quitar datos nulos de Rainfall
moda_RainToday_yes = muestra_o_sucia['Rainfall'][(muestra_o["RainToday"] == "Yes")].mode()
moda_RainToday_no = muestra_o_sucia['Rainfall'][(muestra_o["RainToday"] == "No")].mode()

def validarNumeroEntero(valor):
  try:
      int(valor)
      it_is = True
  except ValueError:
      it_is = False
  return it_is

def validarNumeroFlotante(valor):
  try:
      float(valor)
      it_is = True
  except ValueError:
      it_is = False
  return it_is

for i in muestra_o_sucia.index:
  if(validarNumeroEntero(muestra_o_sucia['Rainfall'][i]) == False & validarNumeroFlotante(muestra_o_sucia['Rainfall'][i]) == False):
    if(muestra_o_sucia['RainToday'][i] == 'Yes'):
      muestra_o_sucia['Rainfall'][i] = moda_RainToday_yes
    else:
      muestra_o_sucia['Rainfall'][i] = moda_RainToday_no
print("\033[4;32m"+'Los datos nulos de Rainfall han sido imputados. DataFrame Actualizado!')
```

```python
imputer = KNNImputer(n_neighbors=3, weights="uniform")
locaciones_muestra = muestra_o_sucia['Location'].unique()

df_numeros = pd.DataFrame()
df_objetos = pd.DataFrame()
muestra_sin_nulos = pd.DataFrame()

nombres_columnas_numericas = muestra_o_sucia.select_dtypes(include=['float64', 'int']).columns
for column in locaciones_muestra:
  iteracion_x_locacion = muestra_o_sucia[(muestra_o_sucia["Location"] == column)]
  
  object_x_iteracion = iteracion_x_locacion.select_dtypes(include=['object'])
  knn_x_iteracion = imputer.fit_transform(iteracion_x_locacion.select_dtypes(include=['float64', 'int']))
  numero_x_iteracion = pd.DataFrame(knn_x_iteracion)

  df_objetos=df_objetos.append(object_x_iteracion, ignore_index=True)
  df_numeros=df_numeros.append(numero_x_iteracion, ignore_index=True)

df_numeros.columns = nombres_columnas_numericas
muestra_sin_nulos = pd.concat([df_objetos, df_numeros], axis=1)
print("\033[4;32m"+'Los datos nulos de Temp9am, Temp3pm, Humidity3pm, Humidity9am, MinTemp y MaxTemp han sido imputados. DataFrame Actualizado!')
```

El proceso de imputación se realizó mediante un ciclo for, en donde se verifica que el valor sea nulo y se realiza el procedimiento correspondiente.

Para la columna RainFall, se utilizó una de las medidas de tendencia central, la moda, para la imputación de estos valores, en donde se consiguió la moda en cuanto a Yes y No. Posteriormente se realizan dos funciones para validar tanto números enteros como flotantes y finalmente se realiza un ciclo for, en donde se verifica que el valor sea nulo (en el caso de que ambas funciones den como resultado False), si es el caso, finalmente se verifica si RainToday es Yes o No para imputar uno u otro.

Para el resto de columnas con valores nulos, se utilizó KNN imputer, en donde primeramente se establecen en una variable las locaciones de la muestra sin repetirse, posteriormente se establecen 3 variables, en donde almacenaremos los datos numéricos, los objetos y finalmente la muestra final sin nulos. Se establece igualmente una variable para almacenar los nombres de las columnas numéricas, ya que al realizar el KNN imputer estos se pierden. Al momento de realizar el ciclo for, se realiza la iteración en base a la localización que corresponda en esa misma, se almacenan los objetos, ya que de lo contrario estos perderían su orden y los datos se terminarían mezclando. ##3.2. Tratamiento de Outliers

Para los valores atípicos la primera herramienta que vamos a utilizar serán los boxplot, ya que nos permite representar cómo están repartidos los datos del conjunto mediante cuartiles, pudiendo destacar valores que están fuera de los bigotes clasificados como valores atípicos del conjunto de datos. Luego mediante una función que recorra cada una de las columnas numéricas de la muestra se imprimirá en pantalla información  importante para el estudio, como el primer cuartil y el tercer cuartil que sirven para calcular el rango intercuartil que vendrían a representar el 50% de los datos, y los bigotes representan el otro 50% (25% bigote superior y 25% el inferior), y por último se calcularon los valores máximos y mínimos del conjunto de datos y se finalizó calculando los valores que son menores al bigote inferior y mayores al superior para encontrar los valores exactos de los outliers.

```python
muestra_sin_nulos.boxplot(return_type='dict', figsize=(24,10))
plt.show()
```

```python
columnas_num = (muestra_sin_nulos.select_dtypes(include=['float64','int']).columns)

for column in columnas_num:
  
  print('╔══════════',column,'══════════╗')
  Q1 = muestra_sin_nulos[column].quantile(.25)
  print(' Primer cuartil: {0:.1f}'.format(Q1), '°C')

  Q3 = muestra_sin_nulos[column].quantile(.75)
  print(' Tercer cuartil: {0:.1f}'.format(Q3), '°C')

  IQR = Q3 - Q1
  print(' Rango inter cuartil: {0:.1f}'.format(IQR), '°C')

  mediana =  muestra_sin_nulos[column].median()
  print(' Mediana: {0:.1f}'.format(mediana), '°C')

  valor_minimo =  muestra_sin_nulos[column].min()
  print(' Valor mínimo: {0:.1f}'.format(valor_minimo), '°C')

  valor_maximo =  muestra_sin_nulos[column].max()
  print(' Valor máximo: {0:.1f}'.format(valor_maximo), '°C')

  bi_calculado = (Q1 - 1.5 * IQR)
  print(' Bigote inferior: {0:.1f}'.format(bi_calculado), '°C')
  
  bs_calculado = (Q3 + 1.5 * IQR)
  print(' Bigote superior: {0:.1f} '.format(bs_calculado), '°C')
  print(' ')
  outliers_detectados = (muestra_sin_nulos[column] < bi_calculado) | (muestra_sin_nulos[column] > bs_calculado)
  outliers = muestra_sin_nulos[column][outliers_detectados]
  print('═ ･ω･ Outliers detectados ･ω･ ═')
  print(' ')
  print(outliers)
  print('╚══════════',column,'══════════╝')
  print(' ')
```
