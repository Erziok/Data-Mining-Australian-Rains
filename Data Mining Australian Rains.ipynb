##1.2 Investigacin (contextualizacin)

Para entender mucho mejor el caso, tuvimos que indagar e investigar cmo funcionan las precipitaciones, ya que no podemos predecir algo de lo que no se tiene conocimiento. Por esto hemos decidido realizar esta investigacin que detalla los siguientes puntos:

* **Composicin de las precipitaciones:**

>Las nubes se componen de hielo o micro gotas de agua dependiendo de la altura de la nube y la temperatura atmosfrica.
>
>El vapor de agua que contiene el aire al enfriarse se condensa, formando las pequeas gotas de agua que forman las nubes. Son tan livianas molecularmente que flotan en el aire (menos densas que el aire).

* **Medicin de las precipitaciones:**

>Precipitacin pluvial: Precipitacin intensa pero de corta duracin.
La forma ms comn de medirla es en MM que sera el espesor de la lmina de agua que se formara a causa de la precipitacin sobre una superficie plana e impermeable y que equivale a litros de agua por metro cuadrado del terreno (L/m2).

* **Origen de la precipitacin:**

>Toda precipitacin se produce por la condensacin del vapor de agua contenido en las masas de aire que se originan cuando dichas masas de aire son forzadas a elevarse y enfriarse (cuando el aire caliente y hmedo cercano a la superficie terrestre asciende y su temperatura baja). Para que se produzca la condensacin es preciso que el aire se encuentre saturado de humedad y existan ncleos de condensacin.
>
>(El aire se satura si contiene muchas molculas de vapor de agua). El estado de saturacin se alcanza normalmente por el enfriamiento del aire. ya que el aire fro se satura con menor cantidad de vapor de agua que el aire caliente.
>
>Por ejemplo, un milmetro cbico (m) de aire a 25C, cuyo contenido de vapor de agua sea de 11g, no est saturado; pero los 11g lo saturan a 10C, y entonces la condensacin (cambio de estado gaseoso a lquido) ya es posible 
>
>La pequeez de las gotas y de los cristales les permite quedar en suspensin en el aire y ser desplazadas por los vientos. Se pueden contar 500 por cm y, sin embargo, 1 m de nube apenas contiene tres gramos de agua.

* **Variacin temporal:**

>Variacin de acuerdo a ciertos ciclos temporales determinados por los movimientos de rotacin y traslacin. Por ejemplo, en San Francisco, California (Estados Unidos), los meses de mayores precipitaciones se dan entre noviembre y marzo, mientras que en Miami, Florida los meses de mayor precipitacin son de mayo a octubre.

* **Variacin de la precipitacin:**

>La distribucin espacial de la precipitacin sobre los continentes es muy variada, as existen extensas reas como los desiertos, donde las precipitaciones son extremadamente escasas, del orden de 0 a 200 mm de precipitacin por ao. En el desierto del Sahara la media anual de lluvia es de apenas algunos mm, mientras que en las reas prximas al golfo de Darin entre Colombia y Panam, la precipitacin anual es superior a 3000 mm, con un mximo de unos 10 metros (10.000 mm). El desierto de Atacama en el norte de Chile, es el rea ms seca de todos los continentes.
>
>De esto quisimos identificar los principales tipos de lluvias, del cual podemos destacar cuatro tipos: 

* **Precipitaciones por presin:**

>Cuando se junta una corriente de aire fro y clido se crea en medio un rea de bajas presiones, y la nubosidad se crea en reas pequeas, provocando que las micro gotas de agua aumenten su densidad ms rpidamente.

* **Precipitaciones por conveccin:**

>Cuando la superficie de la tierra est muy caliente y hay viento fro en la troposfera, el aire clido que pesa menos, asciende y al encontrarse con el aire fro produce cumulonimbos que dan lugar a lluvia acompaadas de tormenta (no suelen durar mucho). 

* **Precipitaciones por orografa:**

>Cuando el aire hmedo llega a las montaas debe elevarse ms y eso provoca que se enfre ms, lo que provoca una condensacin del vapor y formacin de las nubes, si la masa aumenta mucho o se unen muchas masas pueden dar sitio a lluvias o nieve.

* **Precipitaciones por frente clido:**

>Si se encuentran dos masas de aire, clido y fro, la masa clida se desplaza por encima de la fra, el lmite de las dos masas da sitio a condensacin y precipitaciones.

##3.2. Tratamiento de Outliers

Para los valores atpicos la primera herramienta que vamos a utilizar sern los boxplot, ya que nos permite representar cmo estn repartidos los datos del conjunto mediante cuartiles, pudiendo destacar valores que estn fuera de los bigotes clasificados como valores atpicos del conjunto de datos. Luego mediante una funcin que recorra cada una de las columnas numricas de la muestra se imprimir en pantalla informacin  importante para el estudio, como el primer cuartil y el tercer cuartil que sirven para calcular el rango intercuartil que vendran a representar el 50% de los datos, y los bigotes representan el otro 50% (25% bigote superior y 25% el inferior), y por ltimo se calcularon los valores mximos y mnimos del conjunto de datos y se finaliz calculando los valores que son menores al bigote inferior y mayores al superior para encontrar los valores exactos de los outliers.

```python
muestra_sin_nulos.boxplot(return_type='dict', figsize=(24,10))
plt.show()
```

```python
columnas_num = (muestra_sin_nulos.select_dtypes(include=['float64','int']).columns)

for column in columnas_num:
  
  print('',column,'')
  Q1 = muestra_sin_nulos[column].quantile(.25)
  print(' Primer cuartil: {0:.1f}'.format(Q1), 'C')

  Q3 = muestra_sin_nulos[column].quantile(.75)
  print(' Tercer cuartil: {0:.1f}'.format(Q3), 'C')

  IQR = Q3 - Q1
  print(' Rango inter cuartil: {0:.1f}'.format(IQR), 'C')

  mediana =  muestra_sin_nulos[column].median()
  print(' Mediana: {0:.1f}'.format(mediana), 'C')

  valor_minimo =  muestra_sin_nulos[column].min()
  print(' Valor mnimo: {0:.1f}'.format(valor_minimo), 'C')

  valor_maximo =  muestra_sin_nulos[column].max()
  print(' Valor mximo: {0:.1f}'.format(valor_maximo), 'C')

  bi_calculado = (Q1 - 1.5 * IQR)
  print(' Bigote inferior: {0:.1f}'.format(bi_calculado), 'C')
  
  bs_calculado = (Q3 + 1.5 * IQR)
  print(' Bigote superior: {0:.1f} '.format(bs_calculado), 'C')
  print(' ')
  outliers_detectados = (muestra_sin_nulos[column] < bi_calculado) | (muestra_sin_nulos[column] > bs_calculado)
  outliers = muestra_sin_nulos[column][outliers_detectados]
  print('  Outliers detectados  ')
  print(' ')
  print(outliers)
  print('',column,'')
  print(' ')
```
